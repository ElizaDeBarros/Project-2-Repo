```python
import zipfile
import pandas as pd
import os

csv_file = "attacks.csv"
#Reads in CSV file into pandas dataframe.
df = pd.read_csv(f"Resources/{csv_file}", sep=',', encoding='latin-1')
#prints the total rows of the dataframe.
print(f"This table started with {len(df.index)} rows.")

#### 'Case Number', 'Year', 'Country', 'Fatal (Y/N)', 'Time', need to not be null
df = df[df['Case Number'].notna()]
df = df[df['Year'].notna()]
df = df[df['Country'].notna()]
df = df[df['Fatal (Y/N)'].notna()]
df = df[df['Time'].notna()]


df = df.loc[df['Time'] != 'X']
df = df.loc[(df['Fatal (Y/N)'] == 'Y') | (df['Fatal (Y/N)'] == 'N')]
df1 = df.loc[(df['Year'] != 0.0) & (df['Year'] >= 1980)]

#prints the row count after null values have been dropped.
print(f"After dropping null values there are {len(df1.index)} rows.")

df1
```
